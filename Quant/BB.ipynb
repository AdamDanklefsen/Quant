```python
from openbb import obb
import yfinance as yf
import pandas as pd

TICKER = "AAPL"

earnings = obb.equity.fundamental.filings(symbol=TICKER, provider="sec",
                                          form_type=["10-Q", "10-K"], limit=5).to_df().T

print(earnings.index)
# print(earnings[['report_date','filing_date']])


balance = obb.equity.fundamental.balance(start_date="2020-01-01", end_date="2020-12-31", symbol=TICKER, period="quarter").to_df().T
print(balance.index)

income = obb.equity.fundamental.income(start_date="2020-01-01", end_date="2020-12-31", symbol=TICKER, period="quarter").to_df().T
print(income.index)
```
